```
import matplotlib.pyplot as plt
import numpy as np
```

```
x=np.array([1,2,3,4,5])
y=x**2+3
print(x)
print(y)
```

```
plt.style.available
```

```
plt.style.use('seaborn')
plt.plot(x,y,label="Current Growth")
plt.plot(x,x**3,label="Expected Growth")
plt.title('My plot')
plt.xlabel('Xaxix')
plt.ylabel('Yaxix')
plt.legend()
plt.show()
```

```
plt.style.use('fast')
plt.scatter(x,y)
```
